# Here is an example of how can we use a corpora and stem that document:

```python
import nltk
nltk.corpus.gutenberg.fileids()
```

```python
# we will be using "melville-moby_dick.txt" file for stemming
text_file=nltk.corpus.gutenberg.words('melville-moby_dick.txt')
my_lines_list=[]
for line in text_file:
    my_lines_list.append(line)
my_lines_list
```
